```python
def fac(n):
    if n==1:
        return 1
    else:
        return n*fac(n-1)
```

```python
fac(5)
```

```python
def fibo(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
```

```python
fibo(7)
```

```python
dic = {1 : 0, # 메모리 기법
       2: 1}

def fibo(n):
    if n in dic.keys():
        return dic[n]
    else:
        dic[n] = fibo(n-1)+fibo(n-2)
        return dic[n]
```

```python
def print_max(a,b,c):
    if a>b:
        if a>c:
            return a
        else:
            return c
    else:
        if b>c:
            return b
        else:
            return c
```

```python
print_max(3,2,1)
```

```python
def print_avg(a):
    b=0
    for i in a:
        b+=i
    return b/len(a)
```

```python
print_avg([1,2,3,4,5,6,7,8,9])
```
