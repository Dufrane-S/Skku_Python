```python
#크롤링

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from tqdm import tqdm # for문 진행상황 확인

def return_numbers(html):
    numbers = html.find('div',class_='lottonum').text.split()
    del numbers[-2]
    numbers = list(map(int,numbers))
    return numbers

#현재 회차를 가져오기 위한 url
url = requests.get("https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EB%A1%9C%EB%98%90")

#url 문자열 정보를 html로 변환
html = BeautifulSoup(url.text)

current = int(html.find('span',class_='f_red').text.replace('회',''))
current


total=[]
#1회부터 30회까지 가져오기
for n in tqdm(range(1,31)):
    url = requests.get(f'https://search.daum.net/search?w=tot&DA=LOT&rtmaxcoll=LOT&&q={n}%ED%9A%8C%EC%B0%A8%20%EB%A1%9C%EB%98%90')
    html = BeautifulSoup(url.text)

    box = return_numbers(html)
    total.append(box)
```

```python
df = pd.DataFrame(total, columns =['1번째','2번째','3번째','4번째','5번째','6번째','보너스'])
df.to_excel('result.xlsx')
df
```

```python
#연금 복권
#회차 가져오기

def return_numbersYG(html):
    numbers = html.find('div',class_='inner').text.split()
    numbers = list(map(int,numbers))
    jo = html.find('td',class_='each').text.split()
    result = jo + numbers
    return result
result=[]
for n in range(1,10):
    url = requests.get(f"https://search.daum.net/search?w=tot&DA=JIM&rtmaxcoll=JIM&&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C%20720%2B%20{n}%ED%9A%8C%EC%B0%A8") 
    html = BeautifulSoup(url.text)
    box = return_numbersYG(html)
    result.append(box)
    print(box)
    time.sleep(1)
```

```python
yg = pd.DataFrame(result, columns =['조','1번째','2번째','3번째','4번째','5번째','6번째'])
yg.to_excel('resultYG.xlsx')
yg
```

```python
url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1')
html = BeautifulSoup(url.text)
```

```python

total=[]
def return_table(html):
    table = html.find('table',class_='type_2')
    table = pd.read_html(str(table))[0] #문자열을 받는 함수 read_html table은 html
    del table['N']
    del table['토론실']
    table = table[table['종목명'].notnull()]
    return table

##최신 페이지 접근
url=requests.get('https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1')
html = BeautifulSoup(url.text)

kospi_page = int(html.find('td',class_='pgRR').find('a')['href'][-2:])


for n in tqdm(range(1,kospi_page+1)):
    url = requests.get(f'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={n}')
    html = BeautifulSoup(url.text)
    total.append(return_table(html))
    time.sleep(1)
total
```

```python
kospi=pd.concat(total,ignore_index=True) #여러개의 데이터 프레임을 하나의 데이터 프레임으로 합쳐주는 기능
kospi.to_excel('kospi.xlsx')
```
