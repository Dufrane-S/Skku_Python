```python
score['평균'] = score.iloc[:,2:7].mean(axis=1)
score
```

```python
score['평균'].rank(ascending = False) # 높은 평균이 1등
```

```python
#3위가 2개일 때, 3.5등으로 순위 부여
score['순위']=score['평균'].rank(ascending = False)
score
```

```python
#3위가 두 개 일때,  3위라고 부여 다음순위 5위
score['순위']=score['평균'].rank(ascending = False,method='min')
score
```

```python
#3위가 두 개 일때, 4위라고 부여 다음순위 5위
score['순위']=score['평균'].rank(ascending = False,method='max')
score
```

```python
#3위가 두 개 일때, 처음 발견된 것은 3위 다음 발견된 것은 4위라고 부여 다음순위 5위
score['순위']=score['평균'].rank(ascending = False,method='first')
score
```

```python
#3위가 두 개 일때, 둘 다 3 위라고 부여 다음순위 4위
score['순위']=score['평균'].rank(ascending = False,method='dense')
score
```

```python
score['MAX'] = score.iloc[:,2:7].max(axis=1)
score
```

```python
score['MIN'] = score.iloc[:,2:7].min(axis=1)
score
```

```python
import numpy as np
x1=np.array([2,4,6,8]) #행렬로 변환 리스트 에는 숫자를 더할 수 없어서 변환
x2=np.array([1,2,1,3])
y=np.array([30,75,65,95])
a1=0
a2=0
b=0
rr=0.005
# x1: 문제 y : 답

for i in range(3000):
    y_hat=a1*x1+a2*x2+b
    error=y_hat-y
    a1_diff = sum(2*error*x1)
    a2_diff = sum(2*error*x2)
    b_diff = sum(2* error * 1)

    a1=a1-a1_diff*rr
    a2=a2-a2_diff*rr
    b=b-b_diff*rr
    print('{}회 학습 // 특성 1기울기 : {} // 특성2 기울기 {} // 절편 : {}'.format(i+1,a1,a2 ,b))
```

```python
plt.scatter(x1,y) # 경사 하강법(Gradient Descent)
plt.plot(x1,a1*x1+a2*x2+b,'r')
```

```python
import pickle

f= open('fish.dat','br')
fish = pickle.load(f)
f.close()
fish.keys()
```

```python
#생선 도미의 길이와 무게
#생선 빙어의 길이와 무게

plt.scatter(fish['bream_length'],fish['bream_weight'])
plt.scatter(fish['smelt_length'],fish['smelt_weight'])
```

```python
# 지도학습(문제, 정답)
# 도미의 길이, 무게, 빙어으이 길이, 무게 : 문제
#정답 : 도미 or 빙어
length=fish['bream_length'] + fish['smelt_length']
weight=fish['bream_weight'] + fish['smelt_weight']
```

```python
len(fish['bream_length'])
```

```python
len(fish['smelt_length'])
```

```python
len(length)
```

```python
#문제 모양 2차원 행렬 모양
#정답 모양 1차원 행렬 모양 만들어 주기

data=[]
for i in range(49):
    data.append([length[i],weight[i]])
data
```

```python
target = [1]*35 + [0]*14
target
```

```python
data = np.array(data) #2차원 행렬
target = np.array(target) #1차원 행렬
```

```python
# k최근접

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(data,target) #학습
```

```python
model.predict([[5,5]]) #예측    
```

```python
model.predict(data)
```

```python
target
```

```python
model.score(data,target) #평가
```

```python
#데이터를 가지고 있다. (75:25) 75로 학습 24로 평가
#랜덤(비추) 75에 빙어가 없을 확률


#랜덤으로 섞어서 요청한 비율만큼 나누어줌
from sklearn.model_selection import train_test_split
#훈련데이터 문제 시험데이터 문제, 훈련데이터 정답, 시험 데이터 정답
train_input, test_input, train_target, test_target= train_test_split(data,target,test_size=0.2) # 8:2로 분할
```

```python
train_input
```

```python
train_target
```

```python
for n in range(1,40):
    model=KNeighborsClassifier(n_neighbors = n) # 기본 : 5개//
    model.fit(train_input,train_target)
    print(n, model.score(test_input,test_target))
```
