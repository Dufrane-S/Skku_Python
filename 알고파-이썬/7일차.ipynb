```python
f = open('boston.dat','rb')
boston = pickle.load(f)
f.close()
data, target = boston[0],boston[1]
```

```python
data.shape
```

```python
#특성 13개 // 표준화
data = boston[0]
target = boston[1]
```

```python
from sklearn.preprocessing import StandardScaler # 평균, 표준편차
from sklearn.preprocessing import RobustScaler # 중앙값(이상치에 강함, 특이값으로 인해 평균이 이상할 떄 좋음)
from sklearn.preprocessing import MinMaxScaler #최소값 0 최대값 1 이상치에 약함

train_input, test_input,train_target,test_target=train_test_split(data,target)
ss= RobustScaler()
train_scaled = ss.fit_transform(train_input) # train_input의 평균과 표준편차를 찾아 적용해서 train_scaled에 넣기
test_scaled = ss.transform(test_input) # train_input의 평균 + 표준편차로 변환 -> 테스트 해볼 값의 평균을 알면 안되니까

for i in range(1,31):
    model = KNeighborsRegressor(n_neighbors=i,n_jobs=-1)
    model.fit(train_scaled,train_target)
    score=model.score(test_scaled, test_target)
    print(i,score)
```

```python
model = KNeighborsRegressor(n_neighbors=2)
model.fit(train_scaled,train_target)
```

```python
plt.plot(range(len(test_target)),test_target)
plt.plot(range(len(test_target)),model.predict(test_scaled))
```

```python
#머신러닝 : 정형 데이터
#딥러닝: 이미지, 텍스트, 음성
# 인공지능 > 머신러닝> 딥러닝

#머신러닝 : 결정트리 모델 기반 : 효과가 좋다. 표준화가 필요없다. 학습에 대해 설명 가능

wine = pd.read_csv('wine.csv') # 0 화이트와인 1 레드와인
del wine['Unnamed: 0']
wine 

data = wine.iloc[:,:-1].to_numpy()
target = wine.iloc[:,-1].to_numpy()
```
