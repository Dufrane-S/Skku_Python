```python
s=random.randint(1,100)
while True:
    n = int(input('값을 입력하세요 : '))
    if s>n:
        print('up')
    elif s<n:
        print('down')
    elif s==n:
        print('정답')
        break;
```

```python
com = random.sample(range(10),3)
print(com)
strike=0

while True:
    strike = 0
    ball = 0
    me = list(map(int, input('숫자 3개 입력').split()))
    for i in range(3):
        if com[i]==me[i]:
            strike+=1
        elif com[i] in me:
            ball+=1

    print('S B')
    print(strike, ball)
    if strike==3:
        break
```

```python
arr=[4,4,4,3, 3]
answer=[0 for i in range(len(arr))]
answer[0]=arr[0]
test=0
for i in range(1,len(arr)):
    if answer[test]==arr[i]:
        continue
    else:
        answer[test]=arr[i]
        test+=1
while(answer.count(0)!=0):
    answer.remove(0)
print(answer)
```
